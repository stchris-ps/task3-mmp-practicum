```python
import pandas as pd
from sklearn.model_selection import train_test_split
import ensembles as es
```

```python
from wtforms import IntegerField
```

```python
import os
import pickle

from collections import namedtuple
from flask_wtf import FlaskForm
from flask_bootstrap import Bootstrap
from flask import Flask, request, url_for
from flask import render_template, redirect

from wtforms.validators import DataRequired
from wtforms import StringField, SubmitField
from wtforms import SelectField, IntegerField, FileField, TextField
```

```python
from pathlib import Path
import numpy as np
```

```python
app = Flask(__name__, template_folder='html')
app.config['BOOTSTRAP_SERVE_LOCAL'] = True
app.config['SECRET_KEY'] = 'hello'
data_path = './../data'
Bootstrap(app)
messages = []


class data_model:
    alg = ''
    train = ''
    targ = ''


class Message:
    header = ''
    text = ''


class TextForm(FlaskForm):
    text = StringField('Text', validators=[DataRequired()])
    submit = SubmitField('Get Result')

class model_params(FlaskForm):
    select = SelectField('Select model type', choices=[('RandomForest', 'GradientBoosting')], validators=[DataRequired()])
    train = FileField('Choose the file with train set data', validators=[DataRequired()])
    targ = FileField('Choose the file with train set target', validators=[DataRequired()])
    submit_fit = SubmitField('Fit model!')
    #submit_chos = SubmitField('Choise from fitted models')

class choz_mod(FlaskForm):
    title = TextField('Choose your model')
    text = StringField('Text', validators=[DataRequired()])------------------------
    submit_fit = SubmitField('Fit model!')
    submit_chos = SubmitField('Choise from fitted models')


class for_predict(FlaskForm):
    test = FileField('Choose the file with set for prediction', validators=[DataRequired()])
    submit = SubmitField('Predict!')


class Response(FlaskForm):
    score = StringField('Score', validators=[DataRequired()])
    sentiment = StringField('Sentiment', validators=[DataRequired()])
    submit = SubmitField('Try Again')


def analyz_nan(data):
    data = data.drop(columns=['index', 'id', 'date'])
    MISS = data.isna().sum()[data.isna().sum()!=0].index
    for i in MISS:
        data[i] = data[i].fillna(data[i].mean(skipna=True))
        #print(data[i].mean(skipna=True))
    return data


def rand_for(data_name_train, data_name_targ,val=None, **kwargs):
    try:
        train = pd.read_csv(data_name_train)
        targ = pd.read_csv(data_name_targ)
        val_train, val_targ = None, None
        train = analyz_nan(train)
        targ = targ.drop(columns=['index'])
        if not val is None:
            train, val_train, targ, val_targ = train_test_split(train, targ, test_size=val)
        alg = es.RandomForestMSE(n_estimators=500, max_depth=15)
        his = alg.fit(train.to_numpy(), targ.to_numpy().T[0], val_train.to_numpy(), val_targ.to_numpy().T[0])
    except Exception as exc:
        app.logger.info('Exception: {0}'.format(exc))
        #score, sentiment = 0.0, 'unknown'

    return alg, his


def grad_bus(data_name_train, data_name_targ, val=None, **kwargs):
    try:
        train = pd.read_csv(data_name_train)
        targ = pd.read_csv(data_name_targ)
        val_train, val_targ = None, None
        train = analyz_nan(train)
        targ = targ.drop(columns=['index'])
        if not val is None:
            train, val_train, targ, val_targ = train_test_split(train, targ, test_size=val)
        alg = es.GradientBoostingMSE(n_estimators=250, learning_rate=0.1, max_depth=5, feature_subsample_size=int(0.5*train.shape[1]), **kwargs)
        his = alg.fit(train.to_numpy(), targ.to_numpy().T[0], val_train.to_numpy(), val_targ.to_numpy().T[0])
    except Exception as exc:
        app.logger.info('Exception: {0}'.format(exc))
        #score, sentiment = 0.0, 'unknown'
    return alg, his


@app.route('/')
@app.route('/index')
def index():
    return render_template('index.html')


@app.route('/index_js')
def get_index():
    return '<html><center><script>document.write("Hello, i`am Flask Server!")</script></center></html>'


@app.route('/clear_messages', methods=['POST'])
def clear_messages():
    messages.clear()
    return redirect(url_for('prepare_message'))


@app.route('/messages', methods=['GET', 'POST'])
def prepare_message():
    message = Message()

    if request.method == 'POST':
        message.header, message.text = request.form['header'], request.form['text']
        messages.append(message)

        return redirect(url_for('prepare_message'))

    return render_template('messages.html', messages=messages)


@app.route('/predict_page', methods=['GET', 'POST'])
def predict_page():
    try:
        pred_form = for_predict()
        #app.logger.info('On text: {0}'.format(par_form.select.data))
        if pred_form.validate_on_submit(pred_form):
            test = pd.read_csv(pred_form.test.data.filename)
            model = request.args.get('model')
            pr = model.predict(analyz_nan(test))
            pr.to_csv('predictions.csv')
        return render_template('from_form.html', form=pred_form)
    except Exception as exc:
        app.logger.info('Exception: {0}'.format(exc))

@app.route('/model_train', methods=['GET', 'POST'])
def get_text_score():
    try:
        par_form = model_params()
        model = data_model
        if par_form.validate_on_submit():
            #app.logger.info('On text: {0}'.format(par_form.select.data))
            model.alg = par_form.select.data
            model.train = par_form.train.data
            model.targ = par_form.train.data
            model.targ.n
            if model.alg=='RandomForest':
                model = rand_for(model.train, model.targ, 0.2)
            else:
                model = grad_bus(model.train, model.targ, 0.2)
            return redirect(url_for('predict_page', model=model))
        return render_template('from_form.html', form=par_form)
    except Exception as exc:
        app.logger.info('Exception: {0}'.format(exc))
```
